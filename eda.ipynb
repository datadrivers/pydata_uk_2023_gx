# A quick look @ housing data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/Housing.csv")
```

```python
df.head()
```

## Parking lots

```python
sns.countplot(x=df.parking, palette="crest")
plt.show()
```

## Price

```python
min = df.price.min() / 1000000
max = df.price.max() / 1000000
mean = round(df.price.mean() / 1000000, 2)
std = round(df.price.std() / 1000000, 2)
```

```python
sns.displot(data=df, x="price", kind="kde", fill="True")
plt.xlabel(f"Price \n \n Min is {min} M. Max is {max} M. Mean is {mean} M. Std is {std} M.")
plt.show()
```

## Furnishing status

```python
sns.countplot(x=df.furnishingstatus, palette="crest")
plt.show()
```
